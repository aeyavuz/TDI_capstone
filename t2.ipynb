```python
def missing_value(df):
    percent_missing = df.isnull().sum() * 100 / len(df)
    q = list(percent_missing)
    missing_value_df = pd.DataFrame({'Column Name': df.columns, 'Percent Missing % ': q})
    return missing_value_df
```

```python
df = df[df['count_qualified'] > 0.0]
df.shape
```

```python
df["solar_per"] = df["existing_installs_count"] / df["count_qualified"] *100
```

```python
set(df.state_name.unique())
```

```python
states_train =['Massachusetts', 'New York', 'New Jersey', 'Rhode Island', 'Maryland', 'Connecticut', 'California']
# selected based on Policy score on solarpowerrocks.com
df_train = df[df.state_name.isin(states_train)]
df_test = df[~df.state_name.isin(states_train)]
print(df.shape)
print(df_train.shape)
print(df_test.shape)
```

```python
X = df[df.columns[0:22]]
X.columns
```

```python
from sklearn import linear_model
# without any selection
X = df[df.columns[1:22]]
y = df[df.columns[22]]
print("X shape", X.shape)
print("y shape", y.shape)
lm = linear_model.LinearRegression()
model = lm.fit(X,y)
predictions = lm.predict(X)
lm.score(X,y)



```

```python
#on train data
X = df_train[df_train.columns[1:22]]
y = df_train[df_train.columns[22]]
print("X shape", X.shape)
print("y shape", y.shape)
lm2 = linear_model.LinearRegression()
model = lm2.fit(X,y)
predictions = lm2.predict(X)
print("lm2 score",lm2.score(X,y))
```

```python
X_test = df_test[df_test.columns[1:22]]
y_test = df_test[df_test.columns[22]]
print("X shape", X_test.shape)
print("y shape", y_test.shape)

predictions_2 = lm2.predict(X_test)
print("lm2 score",lm2.score(X_test,y_test))
```
